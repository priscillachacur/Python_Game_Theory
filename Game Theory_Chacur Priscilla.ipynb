# Game Theory - The Python Equilibrium


<span class="mark">Description of the game theory situation</span>

There are 2 drivers both in separate cars and in different and opposite directions but on the same road.
Unfortunately, due to construction on the road, only one car can go through at once, therefore the drivers have to make a decision on whether to continue driving and risking for a collision or to stop and wait for the other car to pass by. 

In this scenario, both drivers want to go through the road as soon as possible regardless of the other driver's choice. If both give in and decides to wait, there is no car collision. There is also the possibility that one of the drivers waits and the others crosses the road, or the other way around. In this case, the player who is able to cross the road is benefited. Finally, the last possibility is that both of them decides not to wait, and both cross at the same time resulting in a car collision.

The drivers decisions are completely independent of one another and one does not know what the other one will decide to do. If they both decide to wait, they are both safe. However, if no one decides to wait, they collide. Otherwise, one benefits from crossing the road faster, and the other one waits.

```python
####### Assignment 1 #######


# import random package
import random 

# Declare input as variable to input user name
user_name   = input(prompt = "What's your name?").capitalize()
playagain=0


#While loop to ensure that the player has the option to play again
while True:
    # Ask if the user wants to play game theory
    if playagain==0:
         user_play   = input("Do you want to play a Game Theory Situation?(Yes/No)").lower()
    # Ask if the user wants to play again
    else:
         user_play   = input("Do you want to play a Game Theory Situation Again?(Yes/No)").lower()
    #Ensuring that the user chooses only acceptable answers
    while user_play != "yes" and user_play != "no":
        user_play = input("""
        Sorry, please choose appropriate answer. 
        Do you want to play a Game Theory Situation?(Yes/No)
        Select:
        - 'Yes' to Play
        - 'No' to Not play
        """).lower()

    print("--"*50)

    # print sentence to welcome player
    print(f"Hello {user_name}, Welcome to Game Theory!")

    #Create a point system with the pay-off matrix regardless of user play choice (the matrix has to be the same dimensions of the dictionary)
    game_points = [[0, 1], [1, -1]]

    #Save game points and results in a dictionary. W: waits, D: Drives (works for user vs. python and python vs. python)
    dictionary_results ={"(W, W)": (game_points[0][0], game_points[0][0]),
                         "(W, D)": (game_points[0][0], game_points[1][0]),
                         "(D, W)": (game_points[0][1], game_points[0][0]),
                         "(D, D)": (game_points[1][1], game_points[1][1])}

    ####### User vs. Python #####
    playagain=1
    #Creating the option of the user wanting to play and give the player context
    if "yes" in user_play:

        print(f"""\n Hi {user_name}!
         We know you are a good driver, but we can't always be in control of everything while in the road!""")

        print("""
        You are on your way back home and encounter a construction in the middle of the road.
        There is another car coming in your direction and only one car can cross the road at the same time.
        For that reason, these are the possibilities of what can happen in the next seconds:""")

        # Board Game
        print(f"""
        {user_name}, this is you situation, think well before making a decision.
        \n
        {user_name} \ Computer \t\t Wait \t\t\t\t\t  Continue Driving
        --------------\t ---------------------------------------------\t ---------------------------
        Wait\t\t Both wait (0 pts) \t\t\t\t {user_name} Waits (0 pts) Python Drives (1 pts)
        Continue Driving \t {user_name} Drives (1 pts) Python waits (0 pts) \t Python and {user_name} drives (-1 pts both)
        """)

        # prompt the user to select what to do
        user_choose = input("""
        What would you do? Would you decide to wait or to continue driving?
        Select:
        - 'W' to Waits
        - 'D' to Drives
        """).upper()

        #Ensuring that the user chooses only acceptable answers
        while user_choose != "W" and user_choose != "D":
            user_choose = input("""
        Sorry, please choose appropriate answer. 
        What would you do? Would you decide to wait or to continue driving?
        Select:
        - 'W' to Waits
        - 'D' to Drives
        """).upper()


        # Computer (Python) chosee randomly either to wait or drive
        python_choose = random.randint(0,1) # 0: Wait; 1: Drive

        # conditional structure with the outputs of each scenario
        if user_choose == "W" and python_choose == 0:
            user_result = dictionary_results["(W, W)"][0]
            python_result = dictionary_results["(W, W)"][1]
            print(f"""
        You've decided to wait, as well as me!
        You earned {user_result} points.
        I earned {python_result}""")

        elif user_choose == "D" and python_choose == 0:
            user_result = dictionary_results["(D, W)"][0]
            python_result = dictionary_results["(D, W)"][1]
            print(f"""
        You've decided to drive, I decided to wait!
        You earned {user_result} points.
        I earned {python_result}""")

        elif user_choose == "W" and python_choose == 1:
            user_result = dictionary_results["(W, D)"][0]
            python_result = dictionary_results["(W, D)"][1]
            print(f"""
        I decided to wait, You decided to drive!
        You earned {user_result} points.
        I earned {python_result}""")

        elif user_choose == "D" and python_choose == 1:
            user_result = dictionary_results["(D, D)"][0]
            python_result = dictionary_results["(D, D)"][1]
            print(f"""
        We both decided to drive, we collided!
        You earned {user_result} points.
        I earned {python_result}""")

        # Declaring a winner based on independent decisions
        if user_result > python_result:
            print(f"""Congrats, I waited and you drove. You won!""")

        elif user_result == python_result:
            print(f"""Well, we both decided to drive and now we collided! :( """)

        elif user_result <= python_result:
            print(f"""Congrats to me, you waited and I drove. I beat you!""")

    ####### Python vs. Python #####

    #Creating the option when the user does not want to play, so python will play against itself 5 times
    if "no" in user_play:
        print(f"""\n
            Python Player 1 will play against Python Player 2 five times so a winner can be declared!""")

        # Board Game
        print(f"""
        These are all possibilities:
        \n
        Player 1 \ Player 2 \t\t Wait \t\t\t\t  Continue Driving
        --------------\t ----------------------------------------\t ---------------------------
        Wait\t\t Both wait (0) \t\t\t\t\t Player 1 Waits (0) Player 2 Drives (1)
        Continue Driving \t Player 1 Drives (1) Player 2 waits (0) \t Player 1 & Player 2 drives (-1 both)
        """)
    #Defining incremental variable for python player 2 strategy
        i= 0

    #Identifying variables
        python1_result = 0
        python2_result = 0

    # Limiting the game for 5 rounds and determining the strategy for each python player
        while i < 5:
            #strategy for python 1 is randomly decided
            python1 = random.randint(0,1) # 0: Wait; 1: Drive
            #strategy for python 2 is to use the rounds as a beginning starting point. When i is even, it will wait otherwise it will drive
            python2 = i % 2 #remainder of 0: Wait, remainder of 1:Drive 

        # conditional structure with the outputs of each scenario
            if python2 == 0 and python1 == 0:
                python1_result = python1_result + dictionary_results["(W, W)"][0]
                python2_result = python2_result + dictionary_results["(W, W)"][1]
                print(f"""
    Player 1 decided to wait, as well as player 2! \n -------------------
        """)
            elif python2 == 1 and python1 == 0:
                python1_result = python1_result + dictionary_results["(D, W)"][0]
                python2_result = python2_result + dictionary_results["(D, W)"][1]
                print(f"""
    Player 2 decided to drive, Player 1 decided to wait! \n -------------------
        """)
            elif python2 == 0 and python1 == 1:
                player1_result = python1_result + dictionary_results["(W, D)"][0]
                python2_result = python2_result + dictionary_results["(W, D)"][1]
                print(f"""
    Player 2 decided to wait, Player 1 decided to drive! \n -------------------
            """)
            elif python2 == 1 and python1 == 1:
                python1_result = python1_result + dictionary_results["(D, D)"][0]
                python2_result = python2_result + dictionary_results["(D, D)"][1]
                print(f"""
    Both python players decided to drive, they collided! \n -------------------
            """)

            #incrementing the i value by 1 everytime
            i=i+1

    #Printing final score results
        print(f"""
        Python player 2 scored {python1_result} & Python player 1 scored {python2_result}!!""")


        # Declaring a winner based on the results above
        if python1_result < python2_result:
            print(f"""
            Congrats Python 1 won!""")

        elif python1_result == python2_result:
            print(f"""
            Well, both players tie!""")

        elif python1_result > python2_result:
            print(f"""
            Congrats Python 2 won!""")

#The End!
```
